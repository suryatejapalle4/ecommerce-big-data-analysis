```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, sum, count

# Create Spark Session
spark = SparkSession.builder \
    .appName("Ecommerce_Big_Data_Analytics") \
    .getOrCreate()

# Read CSV File
df = spark.read.csv(
    "ecommerce_dataset_updated.csv",
    header=True,
    inferSchema=True
)

# Display Original Columns
print("Original Columns:")
print(df.columns)

# Rename Columns
df = (df
      .withColumnRenamed("Price (Rs.)","Price")
      .withColumnRenamed("Discount (%)","Discount")
      .withColumnRenamed("Final_Price(Rs.)","Final_Price"))

# Display Schema
print("\nSchema")
df.printSchema()

# Show Data
print("\nFirst Five Records")
df.show(5)

# Total Records
print("Total Records =", df.count())

# Total Revenue
print("\nTotal Revenue")
df.select(sum("Final_Price").alias("Revenue")).show()

# Average Price
print("\nAverage Final Price")
df.select(avg("Final_Price").alias("Average Price")).show()

# Category Wise Sales
print("\nCategory Wise Sales")
df.groupBy("Category") \
  .agg(
      count("*").alias("Orders"),
      sum("Final_Price").alias("Revenue")
  ) \
  .show()

# Payment Method Count
print("\nPayment Method Analysis")
df.groupBy("Payment_Method") \
  .count() \
  .show()

# Average Discount
print("\nAverage Discount by Category")
df.groupBy("Category") \
  .agg(avg("Discount").alias("Average Discount")) \
  .show()

# Top 10 Costliest Purchases
print("\nTop 10 Highest Purchases")
df.orderBy(col("Final_Price").desc()) \
  .show(10)

spark.stop()
```
